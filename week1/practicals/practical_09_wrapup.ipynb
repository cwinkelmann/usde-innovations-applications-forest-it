## Further Topics to Explore

### Use DeepFaune
DeepFaune provides a ready-to-use pipeline for species classification in camera trap images without coding. It includes pre-trained models for a variety of species and can be a good starting point for your own projects. You can find it at [deepfaune](https://www.deepfaune.cnrs.fr/en/)

### AddaxAI
Then there is AddaxAI which is another NoCode Product for species classification. It is a bit more user-friendly and has a nice interface. You can find it at [AddaxAI](https://addaxai.com/)
![AddaxAI.png](assets/AddaxAI.png)

### Is a Zebra on this photo the same Zebra as on this other photo?
- **Re-identification** — matching individuals across images, e.g. to track
  animal movements or count unique individuals. Uses metric learning and
  embedding spaces rather than classification. See [RAPID](https://github.com/cwinkelmann/RAPID-animal-reidentification)

### **Habitat Mapping** The segmentation practical was a teaser.
Can you find a correlation between animal presence and habitat features? You would need to have data of animal presence (e.g. from camera traps) and habitat features (e.g. from satellite imagery) to train a model that predicts habitat suitability for the species of interest. This is a common task in conservation biology and can help identify critical habitats for protection. Practical 7 created a glipse into that. CORINE land cover could just work as well. ### Remote Server Setup
When you want to outsource the computations to a remote server, you can use the Server Infrastructure at HNEE. With this you can Train much larger models with much more data.
#### Remote SSH Connection
If you are running on a remote server, set up an SSH connection with port forwarding for JupyterLab. In PyCharm this is pretty much automatic when you set up a remote interpreter. Make sure to sync not to temp directories, but to a persistent location on the server. For this your [VPN needs to be active](https://zammad.hnee.de/help/de-de/30-vpn-laufwerke) and your [ssh public key(https://docs.github.com/en/authentication/connecting-to-github-with-ssh/generating-a-new-ssh-key-and-adding-it-to-the-ssh-agent)] needs to be added to the server by [Jens Mueller](https://www.hnee.de/p/jens-mueller).
Then connect to the remote server
```bash
olive: ssh -i .ssh/id_ed25519 <YOUR-USERNAME>@10.188.2.1
carrot: ssh -i .ssh/id_ed25519 <YOUR-USERNAME>@10.188.1.1
```
There you will get a linux terminal where you can set up the conda environments as described above. When you start JupyterLab, make sure you can connect from you home machine.
